```python
id_collection = []
name_collection = []
index_list = []

for index, row in movies_meta.iterrows():
    collection_info = row["belongs_to_collection"]
    
    if (not pd.isnull(collection_info)):
        
        
        parsed_collection_info = ast.literal_eval(collection_info)
        
        if not isinstance(parsed_collection_info, float):
            index_list.append(index)
            id_collection.append(parsed_collection_info['id'])
            name_collection.append(parsed_collection_info['name'])
```

# Edit Credits Data

```python
credits = pd.read_csv("C:\\Users\\NG\\Documents\\The Movie Dataset\\Data\\credits.csv") 
```

```python
credits.head()
```

```python
id_movie_list =[]
director_name_list = []
lead_role_list = []

for index, row in credits.iterrows():
    id_movie = row['id']
    cast = row['cast']
    crew = row['crew']
    
    parsed_cast = ast.literal_eval(cast)
    parsed_crew = ast.literal_eval(crew)
    
    id_movie_list.append(id_movie)
    
    for index in range(len(parsed_cast)):
        if parsed_cast[index]['order'] == 0:
            lead_role_list.append(parsed_cast[index]['name'])
            break
    
    for index in range(len(parsed_crew)):
        if parsed_crew[index]['job'] == 'Director':
            director_name_list.append(parsed_crew[index]['name'])
            break
    
```

```python
credits_df = pd.DataFrame(list(zip(id_movie_list,director_name_list, lead_role_list)), columns=['id_movie', 'director_name', 'lead_role'])
```

```python
credits_df.head()
```
